```python
def detect_outliers(sr):
    q1 = sr.quantile(0.25)
    q3 = sr.quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    outliers = sr.loc[(sr < fence_low) | (sr > fence_high)]
    return outliers
detect_outliers(data['points'])
```

## 3. 数据缺失的处理
统计所有属性的数据缺失情况：

```python
print(data.isnull().sum(axis=0))
```

### 3.1 处理country属性缺失
缺失原因：统计失误，可以通过属性的相关关系来填补缺失值，查找country属性缺失的葡萄园所在国家并进行填充

```python
import copy
designation2country = {
    "Askitikos":"Greece",
    "Shah":"US",
    "Piedra Feliz":"Chile",
}
data_country = copy.deepcopy(data)
for index,row in data_country.iterrows():
    if not row['country']:
        row['country'] = designation2country[row['designation']]
```

```python
import matplotlib.pyplot as plt
plt.subplot(2,1,1) 
data["country"].value_counts(dropna = False).plot(kind='bar',figsize=(20,8))
plt.subplot(2,1,2) 
data_country["country"].value_counts(dropna = False).plot(kind='bar',figsize=(20,8))
```

### 3.2 处理designation属性缺失
缺失原因：葡萄园可能较小或葡萄来源不明，将缺失部分剔除

```python
data_designation = data.dropna(subset=['designation'])
plt.subplot(2,1,1) 
data["designation"].value_counts(dropna = False)[:100].plot(kind='bar',figsize=(20,8))
plt.subplot(2,1,2) 
data_designation["designation"].value_counts(dropna = False)[:100].plot(kind='bar',figsize=(20,8))
```

### 3.3 处理price属性缺失
缺失原因：通过数据对象之间的相似性来填补缺失值

```python
data_price= copy.deepcopy(data)
fun = lambda x: x.fillna(data_price.groupby('variety').price.mean()[x.variety])
data_price = data_price.apply(lambda x: fun(x), axis=1)

plt.subplot(2,1,1) 
data_price.boxplot(['price'],return_type='dict',vert=False)
plt.subplot(2,1,2) 
data.boxplot(['price'],return_type='dict',vert=False)
```

```python
print(data_price.isnull().sum(axis=0))
```

### 3.4 处理region_1属性缺失
缺失原因：通过相似对象的最高频率来填补缺失值

```python
data_region1= copy.deepcopy(data)
for index,row in data_region1.iterrows():
    if not row['region_1']:
        row['region_1'] = data[data['variety']==row['variety']]['region_1'].mode()
plt.subplot(2,1,1) 
data["region_1"].value_counts(dropna = False)[:50].plot(kind='bar',figsize=(20,8))
plt.subplot(2,1,2) 
data_designation["region_1"].value_counts(dropna = False)[:50].plot(kind='bar',figsize=(20,8))
```

### 3.5 处理region_2属性缺失
缺失原因：将缺失部分剔除

```python
data_region2 = data.dropna(subset=['region_2'])
plt.subplot(2,1,1) 
data["region_2"].value_counts(dropna = False)[:50].plot(kind='bar',figsize=(20,8))
plt.subplot(2,1,2) 
data_region2["region_2"].value_counts(dropna = False)[:50].plot(kind='bar',figsize=(20,8))
```
